## TF-IDF

**TF — term frequency**\
Вес слова в пределах одного документа.\
Отношение числа вхождения некоторого слова к общему числу слов в документе
$$
tf(t, d) = \frac{n_t}{\sum_k{n_k}}
$$
В числителе число вхождений слова $t$ в документ, в знаменателе общее количество слов в документе

**IDF — inverse document frequency**\
Инверсия частоты, с которой некоторое слово встречается в документах коллекции. Учёт IDF уменьшает вес широкоупотребительных слов. \
$$
idf(t, D) = log\frac{|D|}{|\{d_i \in D| t \in d_i\}|}
$$
$|D|$ - Число документов в коллекции\
$|\{d_i \in D| t \in d_i\}|$ - Число документов из коллекции $D$ в которых встречается $t$

$$tf-idf(t, d, D) = tf(t, d) * idf(t, D)$$

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(stop_words='english')
tfidf
```

```python
tfidf_matrix = tfidf.fit_transform(netflix_movies['description'])
```

```python
# Разреженная scipy-матрица
tfidf_matrix
```

## Cos Similarity

**Косинусная сходимость** - мера сходства между двумя векторами, которая используется для измерения косинуса угла между ними.\
косинусное сходство двух документов изменяется в диапазоне от 0 до 1, поскольку частота терма (веса tf-idf) не может быть отрицательной. Угол между двумя векторами частоты терма не может быть больше, чем 90°.

Если даны два вектора признаков, A и B, то косинусное сходство, cos(θ), может быть представлено используя скалярное произведение и норму:

$$
similarity = cos(\theta) = \frac{A*B}{||A|| ||B||} = \frac{\sum_i{A_i*B_i}}{\sqrt{\sum_i{(A_i)^2}}*\sqrt{\sum_i{(B_i)^2}}}
$$

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cos_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
cos_sim
```

```python
cos_sim.shape
```

```python
netflix_movies.shape[0]
```

```python
index = pd.Series(netflix_movies.index, index=netflix_movies['title'])
index
```

```python
cos_sim_df = pd.DataFrame(cos_sim, columns=index.index)
cos_sim_df['title'] = index.index
cos_sim_df = cos_sim_df.set_index('title')
```

```python
cos_sim_df.head()
```

```python
plt.figure(figsize=(20, 20))
sns.heatmap(cos_sim_df);
```

```python
def get_recommendation(title, cos_sim=cos_sim_df):
    ind = index[title]
    sim_scores = list(enumerate(cos_sim_df.loc[title]))

    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:10]

    movie_index = [i[0] for i in sim_scores]
    return netflix_movies['title'].iloc[movie_index]
```
